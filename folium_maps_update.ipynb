```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# Read the CSV file containing latitude, longitude, and postcode data
data = pd.read_csv('final.csv')

# Create a map centered around the mean latitude and longitude coordinates
map_center = [data['latitude'].mean(), data['longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=13)



# Convert latitude and longitude data to a list of points for heatmap
heat_data = data[['latitude', 'longitude']].values.tolist()

# Create a HeatMap layer
HeatMap(heat_data).add_to(mymap)
#######

# Define different map tile options

folium.TileLayer('cartodbpositron').add_to(mymap)
folium.TileLayer('cartodbdark_matter').add_to(mymap)

folium.LayerControl().add_to(mymap)


########
# Save the map to an HTML file
mymap.save('heat_map.html')

# Display the map
mymap
```
