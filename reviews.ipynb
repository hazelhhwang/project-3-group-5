```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('top_rated-res-mel_cbd.csv')

# Drop the 'name' column as it's not suitable for visualization
df = df.drop(columns=['name'])

# Create a pairplot for cuisine type, rating, and number of reviews
sns.pairplot(df, hue='cuisinetype', palette='Set1')
plt.suptitle('Pairplot of Cuisine Type, Rating, and Number of Reviews', y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('cuisine-avg_rating-no.of_rest.csv')

# Assuming you have columns 'cuisine type', 'avg_rating', and 'num_restaurants'
cuisine_types = df['cuisinetype']
avg_ratings = df['avg_rating']
num_restaurants = df['num_restaurants']  # Assuming this column exists in your CSV file

# Create a scatter plot
plt.figure(figsize=(15, 9))
plt.scatter(num_restaurants, avg_ratings, s=100, color='skyblue', alpha=0.9)
plt.title('Number of Restaurants vs. Average Rating by Cuisine Type')
plt.xlabel('Number of Restaurants')
plt.ylabel('Average Rating')

# Annotate each point with cuisine type
for i, cuisine_type in enumerate(cuisine_types):
    plt.annotate(cuisine_type, (num_restaurants[i], avg_ratings[i]), fontsize=10, rotation=47)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('Avg_priceLevel - Cuisinetype.csv')

# Mapping non-numeric values to numeric equivalents
price_level_mapping = {
    '$': 1,
    '$$ - $$$': 2,
    '$$$$': 3
}

# Convert 'avg_price_level' column to numeric using the mapping
df['avg_price_level'] = df['avg_price_level'].map(price_level_mapping)

# Group the data by cuisine type and calculate the average price level
avg_price_level = df.groupby('cuisinetype')['avg_price_level'].mean().reset_index()

# Extracting data
cuisine_types = avg_price_level['cuisinetype']
average_price_level = avg_price_level['avg_price_level']

# Round the average price levels
average_price_level_rounded = average_price_level.round()

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(cuisine_types, average_price_level_rounded, color='skyblue')
plt.title('Average Price Level by Cuisine Type')
plt.xlabel('Cuisine Type')
plt.ylabel('Average Price Level')
plt.yticks([1, 2, 3])  # Set the y-ticks to only display 1, 2, and 3
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into a DataFrame
df = pd.read_csv('top_10-res-max_reviews-max_rating.csv')

# Assuming your CSV file has columns 'restaurant_name' and 'review_score'
restaurant_names = df.iloc[:10, 0]  # Extracting the first 10 restaurant names
review_scores = df.iloc[:10, 3]      # Extracting the review scores for the first 10 restaurants

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(range(len(restaurant_names)), review_scores, color='skyblue')
plt.title('Top 10 Restaurants by Review Score')
plt.xlabel('Restaurant')
plt.ylabel('Review Score')
plt.xticks(range(len(restaurant_names)), restaurant_names, rotation=45, ha='right')  # Set x-ticks to restaurant names
plt.tight_layout()
plt.show()
```
